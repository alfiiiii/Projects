###### Melihat Plot ACF dan PACF Data

```python
data = pd.read_csv("databaru.csv")

# Menetapkan 'Pintu Masuk' sebagai index dan menghapusnya dari data yang akan diproses
data.set_index("Pintu Masuk", inplace=True)

# Transpose DataFrame untuk membuat bulan sebagai index
data = data.T

# Mengatur index menjadi datetime
data.index = pd.date_range(start="2014-01", periods=len(data), freq="M")

# Menggunakan kolom pertama sebagai data time series
series = data.iloc[:, 0]  # Ganti indeks kolom jika ingin menggunakan kolom lain

# Menentukan lags yang sesuai berdasarkan ukuran data
max_lags = min(10, len(series) // 2 - 1)  # Set lags yang lebih kecil dari setengah ukuran data

# Plot ACF dan PACF
fig, ax = plt.subplots(2, 1, figsize=(12, 8))

# Plot ACF
plot_acf(series, lags=max_lags, ax=ax[0])
ax[0].set_title("Autocorrelation Function (ACF)")

# Plot PACF
plot_pacf(series, lags=max_lags, ax=ax[1])
ax[1].set_title("Partial Autocorrelation Function (PACF)")

plt.tight_layout()
plt.show()
```

**Interpretasi Grafik**

Grafik ACF
*   Grafik ACF menunjukkan korelasi antara nilai-nilai dalam time series dengan lag-nya sendiri pada berbagai keterlambatan (lag).
*   Lag 0 memiliki korelasi 1 karena itu adalah korelasi dengan dirinya sendiri.
*   Pada lag 1, terdapat nilai ACF yang signifikan dan positif, tetapi nilai ACF dengan cepat turun dan berada dalam interval kepercayaan (shaded area) untuk lag berikutnya. Ini menunjukkan bahwa hanya ada sedikit korelasi yang signifikan setelah lag 1.
*   Jika ACF menunjukkan penurunan secara eksponensial, ini dapat mengindikasikan adanya komponen moving average (MA) dalam model.

Grafik PACF
*   Grafik PACF menunjukkan korelasi antara variabel dan lag-nya sendiri, setelah mengontrol pengaruh lags lainnya.
*   Lag 1 memiliki nilai PACF yang signifikan, menunjukkan bahwa terdapat korelasi parsial yang kuat antara nilai saat ini dan nilai pada lag 1.
*   PACF pada lag 2 dan seterusnya tidak signifikan, karena berada dalam interval kepercayaan (shaded area), menunjukkan bahwa tidak ada korelasi parsial yang signifikan pada lags tersebut.
*   PACF yang menunjukkan cutoff yang jelas setelah lag 1 mengindikasikan adanya komponen autoregressive (AR) dalam model.

##### Melakukan Prediksi dengan Analisis Regresi

```python
data = pd.read_csv("databaru.csv")

# Menetapkan 'Pintu Masuk' sebagai index dan menghapusnya dari data yang akan diproses
data.set_index("Pintu Masuk", inplace=True)

# Transpose DataFrame untuk membuat bulan sebagai index
data = data.T

# Mengatur index menjadi datetime dengan frekuensi bulanan
data.index = pd.date_range(start="2014-01", periods=len(data), freq='M')

# Reset index untuk analisis regresi
data.reset_index(inplace=True)
data.rename(columns={"index": "Date"}, inplace=True)

# Mengonversi bulan ke format numerik
data["Month"] = data["Date"].dt.month

# Fitur dan target variabel
X = data[["Month"]]
y = data.iloc[:, 1]  # Menggunakan kolom pertama sebagai target variabel

# Membagi data menjadi set pelatihan dan pengujian
X_train = X
y_train = y

# Membuat model regresi linear
model = LinearRegression()
model.fit(X_train, y_train)

# Melakukan prediksi
y_pred = model.predict(X_train)

# Menghitung metrik evaluasi
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)

# Menampilkan hasil
print(f"Mean Squared Error: {mse}")
print(f"R Score: {r2}")

# Visualisasi hasil regresi
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Data Asli")
plt.plot(X, y_pred, color="red", linewidth=2, label="Garis Regresi")
plt.title("Analisis Regresi Linear Jumlah Kunjungan Wisatawan")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Kunjungan")
plt.legend()
plt.grid(True)
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

**Interpretasi Grafik**

Grafik tersebut menampilkan analisis regresi linier yang menggambarkan hubungan antara bulan dan jumlah kunjungan wisatawan sepanjang tahun. Titik-titik biru pada grafik merepresentasikan data asli jumlah kunjungan setiap bulannya, yang tersebar di sekitar garis regresi. Garis regresi merah menunjukkan tren keseluruhan jumlah kunjungan wisatawan, yang tampak sedikit menurun dari Januari hingga Desember. Meskipun ada beberapa bulan, seperti Juni dan Agustus, di mana jumlah kunjungan wisatawan lebih tinggi dari yang diharapkan berdasarkan tren linier, secara umum terlihat adanya penurunan jumlah kunjungan wisatawan selama setahun. Tren penurunan ini menunjukkan bahwa jumlah wisatawan cenderung berkurang seiring berjalannya waktu, meskipun variasi musiman mungkin mempengaruhi fluktuasi angka kunjungan di bulan-bulan tertentu.

###### Melakukan Prediksi 12 Bulan Kedepan

```python
# Membagi data menjadi set pelatihan dan pengujian
X_train = X
y_train = y

# Melakukan prediksi untuk data pelatihan
y_pred_train = model.predict(X_train)

# Membuat data untuk prediksi di masa depan (misalnya, 12 bulan ke depan)
future_dates = pd.date_range(start=data['Date'].max() + pd.DateOffset(months=1), periods=12, freq='M')
future_months = future_dates.month
future_X = pd.DataFrame(future_months, columns=['Month'])

# Melakukan prediksi untuk bulan di masa depan
future_y_pred = model.predict(future_X)

# Data historis
plt.plot(data["Date"], y, label="Data Historis", color="blue")

# Garis regresi untuk data historis
plt.plot(data["Date"], y_pred_train, label="Garis Regresi Historis", color="red")

# Prediksi di masa depan
plt.plot(future_dates, future_y_pred, label="Prediksi Masa Depan", color="green", linestyle="--")

plt.title("Analisis Regresi Linear Jumlah Kunjungan Wisatawan")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Kunjungan")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Interpretasi Grafik**

Gambar tersebut menunjukkan hasil analisis regresi linear terhadap jumlah kunjungan wisatawan dalam periode waktu tertentu. Data historis yang digambarkan dengan garis biru menunjukkan fluktuasi jumlah kunjungan yang signifikan, dengan beberapa puncak dan penurunan yang tajam. Garis regresi historis berwarna merah menggambarkan tren umum dari data historis tersebut, menunjukkan sedikit penurunan dalam jumlah kunjungan wisatawan seiring berjalannya waktu. Prediksi untuk masa depan ditunjukkan oleh garis hijau putus-putus memperkirakan tren penurunan lebih lanjut dalam jumlah kunjungan wisatawan. Hal ini menunjukkan bahwa, berdasarkan data yang ada, jumlah wisatawan diprediksi akan menurun jika tren saat ini terus berlanjut.
